# Sample Python Notebook

Este é um notebook Python básico demonstrando conceitos fundamentais.

## 1. Variáveis e Tipos de Dados

```python
# Variáveis básicas
nome = "Python"
versao = 3.11
ano = 2024
ativo = True

print(f"Linguagem: {nome}")
print(f"Versão: {versao}")
print(f"Ano: {ano}")
print(f"Ativo: {ativo}")
```

## 2. Listas e Loops

```python
# Lista de números
numeros = [1, 2, 3, 4, 5]

# Loop para imprimir cada número
for numero in numeros:
    print(f"Número: {numero}")
```

## 3. Funções

```python
# Função simples
def saudacao(nome):
    return f"Olá, {nome}! Bem-vindo ao Python!"

# Chamando a função
mensagem = saudacao("Estudante")
print(mensagem)
```

## 4. Condicionais

```python
# Verificando condições
idade = 18

if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```

## 5. Dicionários

```python
# Dicionário com informações
pessoa = {
    "nome": "João",
    "idade": 25,
    "cidade": "São Paulo"
}

# Acessando valores
print(f"Nome: {pessoa['nome']}")
print(f"Idade: {pessoa['idade']}")
print(f"Cidade: {pessoa['cidade']}")
```
